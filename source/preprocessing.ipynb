```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# read data file
df = pd.read_csv("yellow_tripdata_2016-01.csv")
df.head()
```

```python
# check data types of columns
df.dtypes
```

```python
# caculate the number of Dataframe rows and columns 
df.shape
```

```python
# select necessary columns
df_selected = df[['VendorID', 'tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'total_amount']]
df_selected.head(5)
```

```python
# create a 1000-row for streaming data
df_stream = df_selected.sample(1000)
df_stream.head(5)
```

```python
# check if df_stream has enough rows
df_stream.count()
```

```python
# create a 1000-row for batch data
df_batch = df_selected.sample(1000)
df_batch
```

```python
# check if df_batch has enough rows
df_batch.count()
```

```python
# export df_stream and df_batch to csv file named 'Streaming_Data.csv' and 'Batch_Data.csv', respectively
df_stream.to_csv("D:\Python\Streaming_Data.csv", header=True, index=False)
df_batch.to_csv("D:\Python\Batch_Data.csv", header=True, index=False)
```
